```python
import os
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = '/data/BRIDGE/OD-CycleGAN/output/log'
fl_list = os.listdir(path)
fl_list
```

```python
tot_loss = p = r = m_ap = {'v8':[], 'v7transform':[], 'bestorg':[], 'v4':[], 'vh+vv':[]}
for i, fl in enumerate(fl_list):
    keyword = fl.split('_')[1][:-4]
    with open(Path(path) / fl_list[i], 'r') as f:
        log = f.readlines()
        for line in log:
            temp = line.strip()
            temp = [el for el in temp.split(' ') if el!='']
            print(float(temp[5]))
            tot_loss[keyword].append(float(temp[5])); p[keyword].append(temp[8]); 
            r[keyword].append(temp[9]); m_ap[keyword].append(temp[10])
```

```python
tot_loss
```

```python
for k, m in tot_loss:
    print(k)
    plt.plot(np.arange(len(m)), m, label='k')
plt.title('Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
# Display the plot
plt.legend(loc='best')
plt.show()
```
