```python
import os
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = '/data/BRIDGE/OD-CycleGAN/output/log'
fl_list = os.listdir(path)
fl_list
```

```python
for i, fl in enumerate(fl_list):
    keyword = fl.split('_')[1][:-4]
    with open(Path(path) / fl_list[i], 'r') as f:
        log = f.readlines()
        for l, line in enumerate(log):
            print(l, line.strip())
```

```python
tot_loss = p = r = m_ap = {'v8':[], 'v7transform':[], 'bestorg':[], 'v4':[], 'vh+vv':[]}
for i, fl in enumerate(fl_list):
    keyword = fl.split('_')[1][:-4]
    with open(Path(path) / fl_list[i], 'r') as f:
        log = f.readlines()
        for line in log:
            temp = line.strip()
            temp = [el for el in temp.split(' ') if el!='']
            #tot_loss[keyword].append(float(temp[5])); 
            #p[keyword].append(float(temp[8])); 
            #r[keyword].append(float(temp[9])); 
            m_ap[keyword].append(float(temp[10])); 
```

```python
for key, values in tot_loss.items():
    plt.plot(np.arange(len(values)), values, label=key)
plt.title('Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
# Display the plot
plt.legend(loc='best')
plt.show()
```

```python
for key, value in p.items():
    print(len(value))
    plt.plot(np.arange(len(value)), value, label=key)
plt.title('Precision')
plt.xlabel('Epochs')
plt.ylabel('Prec')
# Display the plot
plt.legend(loc='best')
plt.show()
```

```python
for key, value in r.items():
    print(len(value))
    plt.plot(np.arange(len(value)), value, label=key)
plt.title('Recall')
plt.xlabel('Epochs')
plt.ylabel('Rec')
# Display the plot
plt.legend(loc='best')
plt.show()
```

```python
for key, value in m_ap.items():
    print(len(value))
    plt.plot(np.arange(len(value)), value, label=key)
plt.title('mAP')
plt.xlabel('Epochs')
plt.ylabel('mAP')
# Display the plot
plt.legend(loc='best')
plt.show()
```
