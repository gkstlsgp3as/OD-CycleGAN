```python
import os
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = '/data/BRIDGE/OD-CycleGAN/output/'
fl_list = os.listdir(path)
fl_list = [f for f in fl_list if f.endswith('txt')]
```

```python
fl_list
```

```python
tot_loss = p = r = m_ap = {'v8':[], 'v4':[], 'v7':[]}
for i, fl in enumerate(fl_list):
    keyword = fl.split('_')[1][:-4]
    with open(Path(path) / fl_list[i], 'r') as f:
        log = f.readlines()
        for line in log:
            temp = line.strip()
            temp = [el for el in temp.split(' ') if el!='']
            #tot_loss[keyword].append(float(temp[5])); 
            #p[keyword].append(float(temp[8])); 
            #r[keyword].append(float(temp[9])); 
            m_ap[keyword].append(float(temp[10])); 
```

```python
for key, values in tot_loss.items():
    plt.plot(np.arange(len(values)), values, label=key)
plt.title('Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
# Display the plot
plt.legend(loc='best')
plt.show()
```
